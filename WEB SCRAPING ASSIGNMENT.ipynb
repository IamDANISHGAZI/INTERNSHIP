# QUESTION 2  Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame



```python
import pandas as pd # importing pandas for making data frame
import numpy as np  
```

```python
p1=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc") # sending request to the url
p1
```

```python
s1=BeautifulSoup(p1.content, 'html.parser') # creating beautiful soup instance
```

```python
nextpage=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
```

```python
 # creating an empty list in which we append both pages movie names 

name1=[]  
for i in s1.find_all('h3',class_="lister-item-header"):
    name1.append(i.text.replace("\n","").split("(")[0].split('.')[1])

# Creating request for next page    
    
nextpage=requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
s2=BeautifulSoup(nextpage.content)
for i in s2.find_all('h3',class_="lister-item-header"):
    name1.append(i.text.replace("\n","").split("(")[0].split('.')[1])
    
# Creating an empty list for storing all Ratings


rating1=[]
for i in s1.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating1.append(i.text.replace('\n',""))
for i in s2.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating1.append(i.text.replace('\n',""))
    
# Creating an empty list for storing all Release year


year1=[]
for i in s1.find_all('span',class_="lister-item-year text-muted unbold"):
    year1.append(i.text.replace('(I)',""))
for i in s2.find_all('span',class_="lister-item-year text-muted unbold"):
    year1.append(i.text.replace('(I)',""))

# Creating dictionary

my_dict={"Movie Name":name1,"Ratings":rating1,"Year of Release":year1}

# By using Pandas we create Data Frame and set the index from 1

df_1=pd.DataFrame(data=my_dict)
df_1.index+=1
df_1
```

# QUESTION 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :i) Headline ii) Time iii) News Link


```python
# First sending request to the url

newsurl=requests.get("https://www.cnbc.com/world/?region=world")
```

```python
# Creating soup 

newssoup=BeautifulSoup(newsurl.content)
```

```python
# Creating empty list to store headings using a tag

heading=[]
for i in newssoup.find_all('a',class_="LatestNews-headline"):
    heading.append(i.text)

# Creating empty list to store time using time tag

time=[]
for i in newssoup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)

# Creating empty list for storing links using div tag and get href links
    
link=[]
for i in newssoup.find_all('div',class_="LatestNews-container"):
    link.append(i.find('a',class_="LatestNews-headline").get("href"))

# Creating dictionary
    
newsdict={"Latest news headline":heading,"Time":time,"News Link":link}


# Creating DataFrame using pandas and representing data with index starting from 1

dfnews=pd.DataFrame(data=newsdict)
dfnews.index+=1
dfnews
```

```python
# Creating soup 

jurlsoup=BeautifulSoup(jurl.content)
```

```python
# Creating empty list for storing titles by using h2 tag

title=[]
for i in jurlsoup.find_all('h2',class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR"):
    title.append(i.text)

# Creating empty list for storing authors using span tag

authors=[]   
for i in jurlsoup.find_all('span',class_="sc-1w3fpd7-0 pgLAT"):
    authors.append(i.text)

# Creating empty list for storing months using span tag

published_month=[]  
for i in jurlsoup.find_all('span',class_="sc-1thf9ly-2 bKddwo"):
    published_month.append(i.text)

# Creating empty list for storing url using li tag

paperurl=[]
for i in jurlsoup.find_all('li',class_="sc-9zxyh7-1 sc-9zxyh7-2 exAXfr jQmQZp"):
    paperurl.append(i.find('a',class_="sc-5smygv-0 nrDZj").get("href"))

# Creating dictionary

journdict={"Paper Title":title,"Author":authors,"Published":published_month,"URL":paperurl}

# Creating DataFrame using pandas and representing data with index starting from 1

dfjourn=pd.DataFrame(data=journdict)
dfjourn.index+=1
dfjourn
```

# 9) Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL


```python
# First sending request to the url

pageurl = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
# Creating soup

pagesoup=BeautifulSoup(pageurl.content)
```

```python
# Creating empty list to store restnames

restname=[]
for i in pagesoup.find_all('a',class_="restnt-name ellipsis"):
    restname.append(i.text)

# Creating empty list to store cuisine 

cuisine=[]
for i in pagesoup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])

# Creating empty list to store location

location=[]
for i in pagesoup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)

# Creating empty list to store Rating

rating=[]
for i in pagesoup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)

# Creating empty list to store images

images=[]
for i in pagesoup.find_all('img',class_="no-img"):
    images.append(i.get('data-src'))
images

# Creating dictionary

restdict={"Restaurant Name":restname,"Cuisine":cuisine,"Location":location,"Rating":rating,"Image url":images}

# Creating DataFrame using pandas and representing data with index starting from 1

resdf=pd.DataFrame(data=restdict)
resdf.index+=1
resdf
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en i) Rank ii) Publication iii) h5-index iv) h5-median

 
 



```python
# Sending Request to url

scpage=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
```

```python
# Creating soup

scsoupe=BeautifulSoup(scpage.content)
```

```python
# Creating empty list for storing ranks 

rank=[]
for i in scsoupe.find_all('td',class_="gsc_mvt_p"):
    rank.append(i.text)
    
# Creating empty list for storing publications

publication=[]
for i in scsoupe.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)

# Creating empty lists for storing all info of h5 index and h5 median
h5=[]
for i in scsoupe.find_all('td',class_="gsc_mvt_n"):
    h5.append(i.text)
h5i=[]
h5m=[]

# Now using for loop and if statement storing h5 index values and h5 median values seperately 

for i in range(0,len(h5)):
    if i%2:
        h5m.append(h5[i])
    else:
        h5i.append(h5[i])

# Creating dictionary

scdict={"Rank":rank,"Publication":publication,"h5-index":h5i,"h5-median":h5m}

# Creating DataFrame using pandas and representing data with index starting from 1

scdf=pd.DataFrame(data=scdict)
scdf.index+=1
scdf
```
